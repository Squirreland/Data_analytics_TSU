gross - True размер заработной платы до вычета всех налогов. False - после вычета

Выбор минимальной зарплаты (from)

```python
def min_salary_fun(salary): # выбор минимальной зарплаты (from)
   dict_salary = ast.literal_eval(salary)
   value = dict_salary.get('from')
   gross = bool(dict_salary.get('gross'))
   if not pd.isna(value):
     if not gross:# нормализуем все зарплаты "до вычета налогов"
       value = round(value*100/87)
   return value
```

Выбор максимальной зарплаты (from)

```python
def max_salary_fun(salary): # выбор максимальной зарплаты (to)
   dict_salary = ast.literal_eval(salary)
   value = dict_salary.get('to')
   gross = dict_salary.get('gross')
   if not pd.isna(value):
     if not gross: # нормализуем все зарплаты "до вычета налогов"
       value = round(value*100/87)
   return value
```

Выбор валюты (currency)

```python
def currency_salary_fun(salary): # выбор валюты
   dict_salary = ast.literal_eval(salary)
   return dict_salary.get('currency')
```

```python
df_new['min_salary']= df_new['salary'].apply(min_salary_fun)
```
